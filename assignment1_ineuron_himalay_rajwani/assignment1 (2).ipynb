# q-3 Write a Python program to accept the user's first and last name and then getting them printed in the the reverse order with a space between first name and last name.


```python
first_name=input("enter the first name")
last_name=input("enter the last name")
print("Name before reversing the string with space is:",first_name+" "+ last_name)
print("Name after reversing the string with space is:",last_name+" "+first_name)
```

```python
print("Name before reversing the string with space is:{}{}{}" .format(first_name," ",last_name))
print("Name after reversing the string with space is: {}{}{}" .format(last_name," ",first_name))
```

```python
l=[first_name,last_name]
st=""
for i in l:
    st=i+" "+st
print(st)
    
    
```
